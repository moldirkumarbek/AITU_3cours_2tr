```python
data = pd.read_csv('dataset_temperature_2.csv')
```

```python
data.rename(columns={"Unnamed: 0": "Time"}, inplace=True)
data = data.set_index('Time')
```

```python
x = data.iloc[:-1, :24]
y = data.iloc[1:, :10]
```

```python
data
```

```python
scaler_x = StandardScaler()
scaler_x.fit(x)
X_trans = scaler_x.transform(x)
X_train, X_test, y_train, y_test = train_test_split(X_trans, y, shuffle=False)
```
